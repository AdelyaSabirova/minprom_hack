```python
import numpy as np
import pandas as pd
import plotly.express as px
from pathlib import Path
import matplotlib.pyplot as plt

# %matplotlib inline
```

```python
exps_src = Path('/home/alex/nkbtech/hdd4/models/minprom_hack')
exp_stats = {}
for exp in exps_src.iterdir():
    exp_stats[exp.name] = pd.read_csv(exp / 'results.csv')
    exp_stats[exp.name].columns = [col.strip() for col in exp_stats[exp.name].columns]
```

```python
fig, axes = plt.subplots(4, 1, figsize=[10,15])
for name, df in exp_stats.items():
    axes[0].plot(df['epoch'], df['metrics/mAP50(B)'], label=name)
    axes[0].set_title('mAP50')
    axes[1].plot(df['epoch'], df['metrics/precision(B)'], label=name)
    axes[1].set_title('precision')
    axes[2].plot(df['epoch'], df['metrics/recall(B)'], label=name)
    axes[2].set_title('recall')
    axes[3].plot(df['epoch'], df['metrics/mAP50-95(B)'], label=name)
    axes[3].set_title('mAP50-95')

for ax in axes:
    ax.legend()
    ax.grid(True)
```
